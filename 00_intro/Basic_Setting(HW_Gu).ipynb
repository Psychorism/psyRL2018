## 1. Anaconda & conda install TensorFlow & Keras
-----

![Anaconda](https://upload.wikimedia.org/wikipedia/en/thumb/c/cd/Anaconda_Logo.png/200px-Anaconda_Logo.png)

*Anaconda* is an **open source distribution** of the Python and R programming languages for data science and machine learning related applications. It contains about a hundred packages including **conda, numpy, Jupyter notebook**, etc.

**Installation** : [Download Anaconda]( https://www.anaconda.com/download/#macos)

+ Check "**Add Anaconda to my PATH Environment variable**"

+ Type **"conda"** on terminal
-----

<img src="https://conda.io/docs/_images/conda_logo.svg" width="200" height="200" />

*Conda* is an **open source package management system** and environment management system. 

**Installation** : included in Anaconda

-----

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/1/11/TensorFlowLogo.svg/440px-TensorFlowLogo.svg.png" width="200" height="200" />

*Tensorflow* is an **open-source software library** for dataflow programming. It is Google Brain's 2nd generation system (following DistBelief).

**Installation** : type **"conda install tensorflow"** on terminal

+ type **'y'** to proceed when asked

-----

<img src="https://upload.wikimedia.org/wikipedia/commons/c/c9/Keras_Logo.jpg" width="200" height="200" />


*Keras* is an **open source neural network library** written in Python, running on top of *Tensorflow*. It is supported in the Tensorflow core library since 2017.

**Installation** : type **"conda install keras"** on terminal

+ need *Tensorflow* to have been installed.

## 3. Git & GitHub
------

*Git* is an open source **distributed version control system**(VCS).

![Git](https://www.edureka.co/blog/wp-content/uploads/2016/12/Staging-Area-Git-Interview-Questions-Edureka-1.png)


First, **make an account** @ [GitHub](https://github.com/)

Second, **download git** @ [Windows](http://git-scm.com/download/win), [Mac](https://sourceforge.net/projects/git-osx-installer/)

Third, **configure the user information** [Guide](http://emflant.tistory.com/123)

    git config --global user.name "Hyunwoo Gu"
    git config --global user.email "...@snu.ac.kr"
    
Fourth, **set folder to be used as a local repo**

    git clone https://github.com/Psychorism/psyRL2018

Fifth, **make some changes** (e.g. copy & paste your .ipynb file)

    git add "~.ipynb"
    git commit -m "first commit"
    
Sixth, **set remote repository**

    git remote add origin https://github.com/Psychorism/psyRL2018
    
Finally, **push your changes**

    git push -u origin master
